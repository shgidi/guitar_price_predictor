```python
# scrape kaggle datsets
response = requests.get('https://www.kaggle.com/datasets.json?sortBy=hottest&group=all&page=1')
data=json.loads(response.content)
total_results = data['totalDatasetListItems']
data=data['datasetListItems']
page=1

for i in range(20, total_results, 20):  # step 20 since we have 20 results per page
    print(f'scraping page {page}')
    page+=1
    url = add_or_replace_parameter(response.url, 'page', page)
    
    urls+=[item['datasetUrl'] for item in data]
    scriptsCount+=[item['scriptCount'] for item in data]
    datasetSize+=[item['datasetSize'] for item in data]
    titles+=[item['title'] for item in data]
    votes+=[item['voteButton']['totalVotes'] for item in data]
    datasetID+=[item['datasetId'] for item in data]
    dates+=[item['dateUpdated'] for item in data]
    response= requests.get(url)
    
    data=json.loads(response.content)['datasetListItems']

    
```
